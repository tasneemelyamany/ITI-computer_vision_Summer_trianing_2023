```python
#question12
def nothing(x):
    pass
cv.namedWindow('colors')
img13 =np.zeros((512, 512, 3),np.uint8)
cv.createTrackbar('Red', 'colors', 0 , 255, nothing)
cv.createTrackbar('Green', 'colors', 0 , 255, nothing)
cv.createTrackbar('Blue', 'colors', 0 , 255, nothing)

while True:
    cv.imshow('colors',img13)
    k=cv.waitKey(1)
    if k==ord('q'):
        break
    red=cv.getTrackbarPos('Red','colors')
    green=cv.getTrackbarPos('Green','colors')
    blue=cv.getTrackbarPos('Blue','colors')
    img13[:] = [blue, green, red]
    print(blue, green, red)
    cv.destroyAllWindows()
```
