```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#question2
img1 = np.zeros((500,500,3), np.uint8)
cv2.namedWindow('line_move')
def nothing(x):
    pass
cv2.createTrackbar('start_x1','line_move',0,499,nothing)
cv2.createTrackbar('end_x1','line_move',0,499,nothing)
cv2.createTrackbar('start_x2','line_move',0,499,nothing)
cv2.createTrackbar('end_x2','line_move',0,499,nothing)
cv2.createTrackbar('blue','line_move',0,255,nothing)
cv2.createTrackbar('red','line_move',0,255,nothing)
cv2.createTrackbar('green','line_move',0,255,nothing)
while True:
    start_x1 = cv2.getTrackbarPos('start_x1', 'line_move')
    end_x1 = cv2.getTrackbarPos('end_x1', 'line_move')
    start_x2 = cv2.getTrackbarPos('start_x2', 'line_move')
    end_x2 = cv2.getTrackbarPos('end_x2', 'line_move')
    red = cv2.getTrackbarPos('red','line_move')
    green=  cv2.getTrackbarPos('green','line_move')
    blue =   cv2.getTrackbarPos('blue','line_move')
    img1 = np.zeros((500,500,3), np.uint8)
    cv2.line(img1,(start_x1,end_x1), (start_x2,end_x2), (blue, green, red), 2)
    cv2.imshow('line_move', img1)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
```

```python
#question3
img2 = np.zeros((500,500,3), np.uint8)
cv2.namedWindow('image')
def nothing(x):
    pass
cv2.createTrackbar('x', 'image', 0 , 499, nothing)
cv2.createTrackbar('y', 'image', 0 , 499 , nothing)
cv2.createTrackbar('r', 'image', 0 , 499, nothing)
cv2.createTrackbar('blue','image',0,255,nothing)
cv2.createTrackbar('red','image',0,255,nothing)
cv2.createTrackbar('green','image',0,255,nothing)
while True:
    x= cv2.getTrackbarPos("x","image")
    y= cv2.getTrackbarPos("y", "image" )
    r = cv2.getTrackbarPos('r', 'image')
    red = cv2.getTrackbarPos('red','image')
    green=  cv2.getTrackbarPos('green','image')
    blue =   cv2.getTrackbarPos('blue','image')
    img2  = np.zeros((500,500,3), np.uint8)
    cv2.circle(img2, (x,y), r, (blue, green, red), 2)
    cv2.imshow('image' , img2)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
#question4 
img3 = np.zeros((800,800,3), np.uint8)
cv2.putText(img3, "Hello", (100,100), cv2.FONT_HERSHEY_DUPLEX, 3, (255,0,0), 3, cv2.LINE_8)
(w , h), c = cv2.getTextSize("Hello", cv2.FONT_HERSHEY_DUPLEX, 3, 3)
cv2.rectangle(img3, (100,100) , (100+w, 100-h) , (255, 255,0), 3)
plt.imshow(img3)
```

```python
#question5
import math
from math import radians
img4 = np.zeros((800,800,3), np.uint8)
cv2.circle(img4,(400,400), 100 , (0,0,255), 3)
cv2.circle(img4,(400,494), 6 , (255,0,0), 3)
cv2.namedWindow('circles')
delta_theta1 = 0
delta_theta2 = 0
theta =  0
while True:
    img4 = np.zeros((800,800,3), np.uint8)
    cv2.circle(img4,(400,400), 100 , (0,0,255), 3)
    x = int(400 + delta_theta1)
    y = int(400 +delta_theta2)
    cv2.circle(img4,(x,y), 6 , (255,0,0), 3)
    theta += 10
    delta_theta1 = 94* math.cos(radians(theta))
    delta_theta2 = 94* math.sin(radians(theta))
    cv2.imshow('circles', img4)
    k = cv2.waitKey(1000)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```
